```python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
```

```python
def step(n):
    out = zeros(n)
    out[int(n/5):] = 1
    return(out)
    
def integrate(x, y0, Ti, dt):
    out = zeros(len(x))
    out[0] = y0
    
    for i in range(1, len(x)):
        out[i] = out[i-1] + dt/Ti * x[i]
    return out

def differentiate(x, Td, dt):
    out = zeros(len(x))
    out[0] = 0
    
    for i in range(1, len(x)):
        out[i] = (out[i]-x[i-1]) * Td/dt
    return out
```

```python
import random
ns = 200
period = 2
out = step(ns)

tVector = linspace(0, period, ns)

tau = [0.05, 0.1, 0.15, 0.2]
#tau = [0.1]
legendLst = []

plot(tVector, out, ":r")

for t in tau:
    integrated = integrate(out, 0, t, period/ns)
    plot(tVector, integrated)
    legendLst.append(t)
    
title("Integrator")
xlabel("Time (s)")
ylabel("Pv [*]")
grid()
legend(["Step"] + legendLst)
```
